```python
from modelscope import AutoModelForCausalLM, AutoTokenizer
device = "cuda" # the device to load the model onto

pth = "/home/admin/data/huggingface_model/mistral/Mistral-7B-Instruct-v0.3"
model = AutoModelForCausalLM.from_pretrained(
    pth,
    torch_dtype="auto",
    device_map="auto",
    # cache_dir="......" #  加载路径
)
tokenizer = AutoTokenizer.from_pretrained(pth)


```

```python
def add_tags_template(tags, question):
    prompt = '''Please generate new tags based on the existing tags and the given task or question. Then, using all the tags, create a new, more challenging version of the task or question.

### Important: The task or question should remain within the same context but with increased difficulty. The new tags should reflect the added complexity.

### Example1:
[Tags]
teamwork, collaboration, project management

[Original Question]
Describe a situation where team collaboration improved the outcome of a project.

Output:
[New Tag]
conflict resolutioncross-functional teams, conflict resolution, leadership under pressure, time-constrained decision making

[New Question]
Describe a situation where a cross-functional team faced significant internal conflicts due to differing priorities and limited resources. Explain how your leadership approach facilitated conflict resolution, effective collaboration, and timely decision making under pressure, ultimately leading to the successful completion of the project within strict time constraints.


### Example2:
[Tags]
time management, productivity, work-life balance

[Original Question]
What strategies can be used to improve time management in a busy work environment?

Output:
[New Tag]
crisis management, high-stakes decision making, multitasking under pressure, long-term sustainability, burnout prevention

[New Question]
In a high-stress, fast-paced work environment where unexpected crises frequently occur, what advanced time management strategies can you implement to effectively balance urgent, high-priority tasks with long-term goals? Additionally, explain how you would prevent burnout and maintain long-term productivity while managing multiple responsibilities under constant pressure.


### Example3:
[Tags]  
problem-solving, critical thinking

[Original Question]  
How do you approach problem-solving in challenging situations?

Output:
[New Tag]  
innovation, creative problem-solving, risk management, proactive decision making

[New Question]  
In a rapidly evolving industry, how do you apply creative problem-solving techniques to address unforeseen challenges while managing risks? Describe a situation where proactive decision-making led to an innovative solution that surpassed expectations.


### Example4:  
[Tags]  
leadership, communication, decision making, delegation

[Original Question]  
How do you make decisions and delegate tasks as a leader?

Output:
[New Tag]  
empathy-driven leadership, conflict mediation, strategic alignment, building high-performance teams

[New Question]  
As a leader, how do you balance empathy-driven decision making with strategic goals while resolving conflicts within the team? Describe how your communication style and delegation strategy helped build a high-performance team aligned with the organization’s long-term vision.


### Example5:  
[Tags]  
customer service, problem-solving, adaptability, conflict resolution, communication

[Original Question]  
How do you handle difficult customers and resolve conflicts effectively?

Output:
[New Tag]  
customer retention, active listening, personalized solutions, crisis communication, de-escalation strategies

[New Question]  
When dealing with highly dissatisfied customers, how do you implement de-escalation strategies and personalized solutions to turn the situation around? Provide an example where your adaptability and crisis communication skills led to long-term customer retention.


### Example6:  
[Tags]  
project management, risk assessment, communication, stakeholder management, timeline management, quality control

[Original Question]  
How do you manage project timelines and ensure quality while handling multiple stakeholders?

Output:
[New Tag]  
strategic resource allocation, crisis mitigation, transparent communication, stakeholder engagement, agile project delivery, continuous improvement

[New Question]  
Describe a time when unforeseen risks jeopardized a project’s timeline. How did you use strategic resource allocation and agile project delivery methods to mitigate crises while maintaining stakeholder engagement? What strategies did you implement to ensure both quality control and continuous improvement throughout the project?


### Example7:  
[Tags]  
teamwork, leadership, adaptability, conflict resolution, time management, strategic planning, decision making

[Original Question]  
How do you balance teamwork, leadership, and time management when leading a project?

Output:
[New Tag]  
collaborative decision making, resilience under pressure, adaptability in crisis, stakeholder alignment, resource optimization, performance tracking, team morale

[New Question]  
In a high-pressure project environment, how do you lead a diverse team through collaborative decision making and adapt to sudden changes? Discuss a situation where you had to balance resource optimization, stakeholder alignment, and performance tracking, while maintaining team morale and ensuring timely delivery of project milestones.


### Your Task:
[Tags]
{tags}

[Original Question]
{question}

Output:
[New Tag]
'''
    return prompt.format(tags=tags, question=question)
```

```python
tags = "command line, CSV, scrolling, file viewing, terminal, text editor."

question = "View tabular file such as CSV from command line, having horizontal and vertical scrolling would be great."


prompt = add_tags_template(tags, question)


input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(device)
output = model.generate(input_ids, max_length=2048, num_return_sequences=1, do_sample=True, temperature=0.8)
print(tokenizer.decode(output[0], skip_special_tokens=True))

```
