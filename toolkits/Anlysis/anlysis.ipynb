```python
import re
from collections import Counter

def clean_text(text):
    # 移除换行符并合并多个空格
    text = re.sub(r'\s+', ' ', text)
    return text

def split_into_sentences(text):
    # 按照句号、感叹号、问号等标点符号拆分句子
    sentences = re.split(r'[.!?]', text)
    # 去除前后空白，并过滤掉空字符串
    sentences = [s.strip() for s in sentences if s.strip()]
    return sentences

def detect_repetitive_text(text, repetition_threshold=0.2):
    cleaned_text = clean_text(text)
    sentences = split_into_sentences(cleaned_text)
    
    # 统计句子的出现频率
    sentence_counts = Counter(sentences)
    total_sentences = len(sentences)
    
    # 找到重复频率超过阈值的句子
    repetitive_sentences = {sentence: count for sentence, count in sentence_counts.items() if count / total_sentences >= repetition_threshold}
    
    if repetitive_sentences:
        print("检测到重复的句子：")
        for sentence, count in repetitive_sentences.items():
            print(f"句子: '{sentence}' - 出现次数: {count}")
    else:
        print("没有检测到大面积的重复。")

# 示例文本
text = """
The gauge on a water tank shows that the tank is 1/3 full of water. To fill the tank, 16 gallons of water are added. After that, another 24 gallons of water are added. The tank is then drained completely. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then, 16 gallons of water are added again. After that, another 24 gallons of water are added. The tank is drained again. Then
"""

# 调用检测函数
detect_repetitive_text(text, repetition_threshold=0.1)
```
